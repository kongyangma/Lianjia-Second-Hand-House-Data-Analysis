```python
df=lianjia.copy()
df['PerPrice']=lianjia['Price']/lianjia['Size']

columns=['Region','District','Garden','Layout','Floor','Year','Size','Elevator','Direction','Renovation','PerPrice','Price']
df=pd.DataFrame(df,columns=columns)
df.head(10)
```

```python
cleaned=df.dropna(how='all')
cleaned.head(10)
```

```python
data=cleaned
```

```python
data.head(10)
```

```python
data_house_count=data.groupby('Region')['Price'].count().sort_values(ascending=False).to_frame().reset_index()
data_house_mean=data.groupby('Region')['PerPrice'].mean().sort_values(ascending=False).to_frame().reset_index()
data_house_count
```

```python
data_house_mean
```

```python
#To display Chinese characters in plots
plt.rcParams['font.sans-serif'] = [u'SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig=plt.figure()
fig,[ax1,ax2,ax3]=plt.subplots(3,1,figsize=(20,15))
sns.barplot(x='Region',y='PerPrice',palette="Blues_d",data=data_house_mean,ax=ax1)
ax1.set_title('Comparision of PerPrice',fontsize=15)
ax1.set_xlabel('Region')
ax1.set_ylabel('Price Per Squaremeter')

sns.barplot(x='Region', y='Price', palette="Greens_d", data=data_house_count, ax=ax2)
ax2.set_title('Comparision of Quantities',fontsize=15)
ax2.set_xlabel('Region')
ax2.set_ylabel('Quantity')

sns.boxplot(x='Region', y='Price', data=data, ax=ax3)
ax3.set_title('Comparision of Prices of Houses',fontsize=15)
ax3.set_xlabel('Region')
ax3.set_ylabel('Price')

plt.show()
```

```python
fig,[ax1,ax2]=plt.subplots(2,1,figsize=(15,10))
sns.distplot(data['Size'],bins=20,ax=ax1,color='r')
sns.kdeplot(data['Size'],shade=True,ax=ax1)

sns.regplot(x='Size',y='Price',data=data,ax=ax2)
plt.show()
```
